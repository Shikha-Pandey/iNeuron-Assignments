Question 1: -
Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input - string = “write write write all the number from from from 1 to 100”
Example output - 5
Explanation - From the given string we can note that the most frequent words are “write” and “from” and
the maximum value of both the values is “write” and its corresponding length is 5

```python
def length_of_highest_freq_word(string):
    count_word = {}
    words = string.split()

    for i in words:
        if i in count_word:
            count_word[i] += 1
        else:
            count_word[i] = 1

    max_frequency = max(count_word.values())
    highest_frequency_words = [word for word, count in count_word.items() if count == max_frequency]

    highest_frequency_word_length = len(highest_frequency_words[0])
    return highest_frequency_word_length

# Test case 1
string = "write write write all the number from from from 1 to 100"
print(length_of_highest_freq_word(string)) 

# Test case 2
string = "The truth is the power"
print(length_of_highest_freq_word(string))

# Test case 3
string = "MAharana was a great leader"
print(length_of_highest_freq_word(string))  
```

Question 2: -
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.
Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }
Example output 1- YES
Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }
Example output 2 - NO

```python
def is_valid_string(s):
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    values = list(char_count.values())
    unique_values = set(values)

    if len(unique_values) == 1:
        return "YES"

    if len(unique_values) > 2:
        return "NO"

    min_count = min(values)
    max_count = max(values)

    if values.count(min_count) == 1 and min_count == 1:
        return "YES"

    if values.count(max_count) == 1 and max_count - min_count == 1:
        return "YES"

    return "NO"

# Test case 1
s = "abc"
print(is_valid_string(s))  

# Test case 2
s = "abcc"
print(is_valid_string(s))  

# Test case 3
s = "aab"
print(is_valid_string(s)) 

# Test case 4
s = "aabbbbcc"
print(is_valid_string(s)) 
```

Question 4. Write a program to download the data from the link given below and then read the data and convert the into
the proper structure and return it as a CSV file.
Link - https://data.nasa.gov/resource/y77d-th95.json
Note - Write code comments wherever needed for code understanding.

```python
import pandas as pd
import requests

def download_data_from_link(url):
    # Send a GET request to download the data from the provided link
    response = requests.get(url)
    
    if response.status_code == 200:
        # Return the response content
        return response.content
    else:
        print("Failed to download data from the provided link.")
        return None

def convert_to_csv(data, output_file):
    # Decode the data from bytes to string
    data_str = data.decode("utf-8")
    
    # Convert the string to a pandas DataFrame
    df = pd.read_json(data_str)
    
    # Save the DataFrame as a CSV file
    df.to_csv(output_file, index=False)

# Provide the link to download the data
link = "https://data.nasa.gov/resource/y77d-th95.json"

# Download the data from the link
data = download_data_from_link(link)

if data is not None:
    # Specify the output file name
    output_file = "data.csv"

    # Convert the data to CSV format and save it to the output file
    convert_to_csv(data, output_file)

    print("Data successfully converted to CSV format and saved to", output_file)
```

```python
import requests
import json

def download_data_from_link(url):
    # Send a GET request to download the data from the provided API link
    response = requests.get(url)
    
    if response.status_code == 200:
        # Return the response content as JSON data
        return response.json()
    else:
        print("Failed to download data from the provided API link.")
        return None

def extract_data(data):
    if data:
        # Extract the show information
        show_info = {
            "Name": data["name"],
            "Language": data["language"],
            "Genres": ", ".join(data["genres"]),
            "Status": data["status"],
            "Premiered": data["premiered"],
            "Official Site": data["officialSite"]
        }
        
        # Extract episode information
        episodes = data["_embedded"]["episodes"]
        episode_info = []
        for episode in episodes:
            episode_info.append({
                "Season": episode["season"],
                "Episode": episode["number"],
                "Name": episode["name"],
                "Airdate": episode["airdate"]
            })
        
        return show_info, episode_info
    else:
        return None

# Provide the API link to download the data
api_link = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"

# Download the data from the API link
data = download_data_from_link(api_link)

if data is not None:
    # Extract the required data
    show_info, episode_info = extract_data(data)
    
    if show_info and episode_info:
        # Print the show information
        print("Show Information:")
        for key, value in show_info.items():
            print(f"{key}: {value}")
        
        print()
        
        # Print the episode information
        print("Episode Information:")
        for episode in episode_info:
            print(f"Season {episode['Season']}, Episode {episode['Episode']}: {episode['Name']}")
            print(f"Airdate: {episode['Airdate']}")
            print()
    else:
        print("Failed to extract data from the downloaded content.")
```

● Get all Pokemons whose spawn time is less than 300 seconds.

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt

# Fetch the data from the URL
url = 'https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json'
response = requests.get(url)
data = response.json()

# Create a DataFrame from the fetched data
df = pd.DataFrame(data['pokemon'])

# Get all Pokemons whose spawn rate is less than 5%
spawn_rate_less_than_5 = df[df['spawn_chance'].notnull() & (df['spawn_chance'] < 0.05)]
print("Pokemons with spawn rate less than 5%:")
print(spawn_rate_less_than_5)

# Plot spawn rate for all Pokemons
plt.figure(figsize=(12, 6))
plt.bar(df['name'], df['spawn_chance'])
plt.xlabel('Pokemon')
plt.ylabel('Spawn Rate')
plt.title('Spawn Rate of Pokemons')
plt.xticks(rotation=90)
plt.show()

# Get all Pokemons that have less than 4 weaknesses
weaknesses_less_than_4 = df[df['weaknesses'].apply(lambda x: isinstance(x, list) and len(x) < 4)]
print("Pokemons with less than 4 weaknesses:")
print(weaknesses_less_than_4)

# Plot weaknesses count for all Pokemons
plt.figure(figsize=(12, 6))
df['num_weaknesses'] = df['weaknesses'].apply(lambda x: len(x) if isinstance(x, list) else 0)
plt.bar(df['name'], df['num_weaknesses'])
plt.xlabel('Pokemon')
plt.ylabel('Number of Weaknesses')
plt.title('Number of Weaknesses of Pokemons')
plt.xticks(rotation=90)
plt.show()

# Get all Pokemons that have no multipliers at all
no_multipliers = df[df['multipliers'].apply(lambda x: isinstance(x, list) and len(x) == 0)]
print("Pokemons with no multipliers:")
print(no_multipliers)

# Get all Pokemons whose spawn time is less than 300 seconds
def convert_spawn_time(time):
    if time is None or time == 'N/A':
        return 0
    minutes, seconds = map(int, time.split(':'))
    return minutes * 60 + seconds

df['spawn_time_seconds'] = df['spawn_time'].apply(convert_spawn_time)
spawn_time_less_than_300 = df[df['spawn_time_seconds'] < 300]
print("Pokemons with spawn time less than 300 seconds:")
print(spawn_time_less_than_300)

# Plot spawn time in seconds for all Pokemons
plt.figure(figsize=(12, 6))
plt.bar(df['name'], df['spawn_time_seconds'])
plt.xlabel('Pokemon')
plt.ylabel('Spawn Time (seconds)')
plt.title('Spawn Time of Pokemons')
plt.xticks(rotation=90)
plt.show()

# Get all Pokemon who have more than two types of capabilities
more_than_2_capabilities = df[df['type'].apply(lambda x: isinstance(x, list) and len(x) > 2)]
print("Pokemons with more than two types of capabilities:")
print(more_than_2_capabilities)
```

Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    
    # Perform POS tagging on the words
    tagged_words = pos_tag(words)
    
    # Initialize counts for each POS category
    counts = {'Noun': 0, 'Verb': 0, 'Pronoun': 0, 'Adjective': 0}
    
    # Count the occurrences of each POS category
    for word, tag in tagged_words:
        if tag.startswith('N'):  # Noun
            counts['Noun'] += 1
        elif tag.startswith('V'):  # Verb
            counts['Verb'] += 1
        elif tag.startswith('PR'):  # Pronoun
            counts['Pronoun'] += 1
        elif tag.startswith('JJ'):  # Adjective
            counts['Adjective'] += 1
    
    return counts

# Test Case 1: Count POS tags in a phrase
phrase = "The quick brown fox jumps over the lazy dog"
result = count_pos_tags(phrase)
print("POS tag counts in the phrase:")
print(result)
print()

# Test Case 2: Count POS tags in a paragraph
paragraph = "I enjoy taking long walks on the beach. It helps me relax and clear my mind. The sound of the waves and the feeling of sand between my toes are soothing. I often find inspiration and solace by the ocean."
result = count_pos_tags(paragraph)
print("POS tag counts in the paragraph:")
print(result)
print()

# Additional Test Case 1: Count POS tags in a question
question = "Can you pass me the red pen, please?"
result = count_pos_tags(question)
print("POS tag counts in the question:")
print(result)
print()

# Additional Test Case 2: Count POS tags in a sentence with contractions
sentence = "I don't like rainy days."
result = count_pos_tags(sentence)
print("POS tag counts in the sentence:")
print(result)
```
